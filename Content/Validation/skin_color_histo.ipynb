```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

from m_estimator import huber_m_estimator

path = r"C:\Users\reinh\Desktop\Synthetic_Video_of_Real_PPG\Content\assets\tex_sample_01.png"
```

```python
img = cv2.imread(path)

red = img[:,:,0].flatten()
green = img[:,:,1].flatten()
blue = img[:,:,2].flatten()

red_mean = np.mean(red)
green_mean = np.mean(green)
blue_mean = np.mean(blue)

red_m = huber_m_estimator(red)
green_m = huber_m_estimator(green)
blue_m = huber_m_estimator(blue)
```

```python
plt.hist(red, bins=256, range=(0, 255))
plt.xlabel("Value")
plt.ylabel("Number")
plt.axvline(red_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {red_mean:.2f}")
plt.axvline(red_m, color='green', linestyle='--', linewidth=2, label=f"M-estimator = {red_m:.2f}")
plt.legend()
plt.show()

plt.hist(green, bins=256, range=(0, 255))
plt.xlabel("Value")
plt.ylabel("Number")
plt.axvline(green_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {green_mean:.2f}")
plt.axvline(green_m, color='green', linestyle='--', linewidth=2, label=f"M-estimator = {green_m:.2f}")
plt.legend()
plt.show()

plt.hist(blue, bins=256, range=(0, 255))
plt.xlabel("Value")
plt.ylabel("Number")
plt.axvline(blue_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {blue_mean:.2f}")
plt.axvline(blue_m, color='green', linestyle='--', linewidth=2, label=f"M-estimator = {blue_m:.2f}")
plt.legend()
plt.show()
```
